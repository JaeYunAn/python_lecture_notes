** 딕셔너리 메서드

* D.get(k)
* D.get(k, v)
* D.keys()
* D.values()
* D.items()
* D.pop(k)
* D.pop(k, v)

.get(key[,default])

* 키에 연결된 값을 반환
* 키에 없으면 None 혹은 기본 값을 반환

```python
person = {'name': 'Alice', 'age': 25}

print(person.get('name'))
print(person.get('country'))
print(person.get('country', 'Unknown'))
print(person['country'])
```

.keys()

* 딕셔너리 키를 모은 객체를 반환

```python
person = {'name': 'Alice', 'age': 25}
print(person.keys())
for i in person.keys():
    print(i)
```

.values()

* 딕셔너리 값을 모은 객체를 반환

```python
person = {'name': 'Alice', 'age': 25}
print(person.values())
for i in person.values():
    print(i)
```

.items()

* 딕셔너리 키/값 쌍을 모은 객체를 반환

```python
person = {'name': 'Alice', 'age': 25}
print(person.items())
for i in person.items():
    print(i)
for key, value in person.items():
    print(key, value)
```

.pop(key[,default])

* 키를 제거하고 연결됐던 값을 반환 (없으면 에러나 default를 반환)

```python
person = {'name': 'Alice', 'age': 25}

print(person.pop('age'))
print(person)
print(person.pop('country', None))
print(person.pop('country'))
```

** defaultdict

* 파이썬 내장 모듈 collections에서 제공하는 딕셔너리의 확장판
* 딕셔너리에 존재하지 않는 키를 조회할 때, 자동으로 '기본값'을 생성해주는 자료구조
* KeyError 걱정 없이 값을 수정하거나 추가할 때 매우 유용함

```python
# 기존: 기본 딕셔너리
a = 'banana'
counts = {}

for i in a:
    if i not in counts:
        counts[i] = 0
    counts[i] += 1

print(counts)
```

```python
# 개선: defaultdict 활용
from collections import defaultdict

a = 'banana'
counts = defaultdict(int)

for i in a:
    counts[i] += 1

print(counts)
```

defaultdict 구문 및 활용 가이드

* 사용법: defaultdict(자료형) 형태로 선언
* 자주 쓰는 패턴
1. 숫자 세기: defaultdict(int) -> 키가 없으면 0으로 초기화
2. 그룹핑/리스트 모으기: defaultdict(list) -> 키가 없으면 빈 리스트[]로 초기화

```python
from collections import defaultdict

fruits = [('red', 'apple'), ('yellow', 'banana'), ('red', 'cherry')]
fruit_by_color = defaultdict(list)

for color, fruit in fruits:
    fruit_by_color[color].append(fruit)

print(fruit_by_color)
```

** 세트 메서드

* s.add(x)
* s.remove(x)

.add(x)

* 세트에 x를 추가

```python
my_set = {'a', 'b', 'c', 1, 2, 3}

my_set.add('d')
print(my_set)
```

.remove(x)

* 세트에서 항목 x를 제거, 항목 x가 없을 경우에는 keyError

```python
my_set = {'a', 'b', 'c', 1, 2, 3}

my_set.remove(2)
print(my_set)

my_set.remove(10)
```

** 세트의 집합 메서드

* set1.difference(set2) -> set1 - set2
* set1.intersection(set2) -> set1 & set2
* set1.issubset(set2) -> set1 <= set2
* set1.issuperset(set2) -> set >= set2
* set1.union(set2) -> set1 | set2
