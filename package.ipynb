패키지
연관된 모듈들을 하나의 디렉토리에 모아놓은 것

사용 목적
모듈들의 이름 공간을 구분하여 충돌을 방지
모듈을 효율적으로 관리하고 할수 있도록 돕는 역할

직접 패키지 만들어보기
* sample.py 파일 생성
* my_package 폴더 생성
* my_package 폴더 내부에 math 폴더와 statistics 폴더 생성
* my_package/math 폴더 내부에 my_math.py 파일 생성 후 다음 코드 작성
* my_package/statistics 폴더 내부에 tools.py 파일 생성 후 다음 코드 작성

```python
# my_package/math/my_math.py
def add(x, y):
    return x + y
```

```python
# my_package/statistics/tools.py
def mod(x, y):
    return x % y
```

직접 만든 패키지 사용하기
* sample.py 파일에 다음 코드 작성해서 실행 결과 확인

```python
#sample.py

from my_package.math import my_math
from my_package.statistics import tools

print(my_math.add(1, 2))    # 출력: 3
print(tools.mod(1, 2))    # 출력: 1
```

라이브러리

라이브러리 > 패키지 > 모듈

패키지: 모듈들을 담은 폴더
라이브러리: 패키지들을 담은 큰 묶음

파이썬 표준 라이브러리
* 파이썬 언어와 함께 제공되는 다양한 모듈과 패키지의 모음
* 다양한 기능이 들어있어 복잡한 작업 쉽게 처리 가능
* math, random, sys (모듈) 및 json, email (패키지) 등 다양한 모듈과 패키지 포함
* 별도 설치 없이 바로 import 해서 사용 가능

파이썬 외부 패키지
* 필요한 기능을 사용하기 위해 직접 설치해서 쓰는 패키지
* 전 세계 개발자들이 만든 다양한 패키지가 존재
* 사용한 패키지를 설치할 때는 pip명령어 사용

pip
외부 패키지들을 설치하도록 도와주는 파이썬의 패키지 관리 시스템

패키지 설치
* 최신 버전 / 특정 버전 / 최소 버전을 명시하여 설치 가능

$ pip install SomePackage
$ pip install SomePackage == 1.0.5
$ pip install SomePackage >= 1.0.4
