공통 시퀀스 메서드

s.index(x)
* 시퀀스 s에서 첫 번째로 일치하는 항목 x의 인덱스를 반환. 없으면, ValueError 발생

s.count(x)
* 시퀀스 s에서 등장하는 항목 x의 개수를 반환

```python
a = 'banana'
print(a.index('n'))

b = [1, 2, 3]
print(b.index(3))
```

```python
a = 'banana'
print(a.count('n'))

b = [1, 2, 3, 3, 3]
print(b.count(3))
```

문자열 탐색 및 검증 메서드

* s.find(x): x의 첫 번째 위치를 반환. 없으면 -1을 반환
* s.isupper(): 문자열 내의 모든 케이스 문자가 대문자인지 확인
* s.islower(): 문자열 내의 모든 케이스 문자가 소문자인지 확인
* s.isalpha(): 문자열의 모든 문자가 알파벳이고 하나 이상의 문자가 포함되어 있으면  True를 반환 (단순 알파펫이 아닌 유니코드 상 letter(한국어도 포함한다.))

```python
print('banana'.find('a'))

print('banana'.find('z'))
```

```python
str1 = 'HELLO'
str2 = 'hello'
print(str1.isupper())
print(str1.islower())
print(str2.isupper())
print(str2.islower())
```

```python
str1 = 'Hello'
str2 = '1251poqifhqwoh'
print(str1.isalpha())
print(str2.isalpha())
```

문자열 조작 메서드(새로운 문자열 반환)

**중요한 것들**
* str.replace(old, new[,count])
* str.strip([chars])

.replace(old, new[,count])

* 기존 문자열에서 'old'라는 부분 문자열이 'new'로 모두 바뀐 문자열의 반환

```python
a = 'Hello, world! world world'
b = a.replace('world', 'python')
c = a.replace('world', 'python', 1)
print(b)
print(c)
```

.strip([chars])

* 선행과 후행 문자가 제거된 문자열의 복사본을 돌려줌
* chars 인자는 제거할 문자 집합을 지정하는 문자열
* 생략되거나 None 이라면, chars 인자의 기본값은 공백을 제거

```python
# 사용자 입력 등에서 불필요한 공백이 포함된 경우
text = '       Hello         World             '

# 1. 아무것도 지정 하지 않으면 '공백(띄어쓰기, 탭, 엔터)'을 제거
clean_text = text.strip()
print(clean_text)
# 문자열 중간의 공백은 지워지지 않는다.

# 2. 제거할 문자 지정하는 경우
a = '!!!!hello world!!!!!'
clean_a = a.strip('!')
print(clean_a)

# 3. 문자열 집합으로 제거(순서 상관 없음)
# 'w', '.', 'c', 'o', 'm' 중 하나라도 양쪽 끝에 있다면 계속 제거
b = 'www.example.com'
print(b.strip('w.com'))
```

.split(sep=None, maxsplit=-1)

* sep를 구분자 문자열로 사용하여 문자열에 있는 단어들의 리스트를 반환
* maxsplit이 주어지면 최대 maxsplit번의 분할이 수행됨
* sep이 지정되지 않거나 None이면, 연속된 공백 문자는 단일한 구분자로 간주하고, 문자열이 선행이나 후행 공백을 포함해도 결과는 시작과 끝에 빈 문자열을 포함하지 않음

```python
# 1. 공백을 기준으로 분리 (기본 동작)
# 여러 개의 공백을 하나로 처리함. 앞, 뒤 공백은 무시
a = '   Hello   Python'
print(a.split())

# 2. 특정 문자를 기준으로 분리
# 지정한 문자를 기준으로 엄격하게 분리함(빈 문자열 발생 가능)
b = '10, 20,, 30'
print(b.split(','))

# 3. 분할 횟수 제한(maxsplit)
# 앞에서부터 1번만 자르고 나머지는 그대로 둔다.
c = 'User/admin/documents'
print(c.split('/', maxsplit=1))
```

.join(iterable)

* iterable의 문자열을 연결한 문자열을 반환

```python
a = ['Python', 'is', 'awesome']

a_1 = ' '.join(a)
a_2 = '-'.join(a)

print(a_1)
print(a_2)
```

** 리스트 전용 리스트 값 추가 및 삭제 메서드

중요한것들
* L.append(x)
* L.extend(iterable)
* L.pop()

.appned(x)

* 리스트 마지막에 항목 x를 추가

```python
a = [1, 2, 3]
a.append(4)
print(a)
```

.extend(iterable)

* 리스트에 다른 반복 가능한 객체의 모든 항목을 추가

** extend()주의사항
* append()와의 비교
a = [1, 2, 3]
a.append([4, 5, 6])
print(a)    # [1, 2, 3, [4, 5, 6]]

* 반복 가능한 객체(iterable)가 아니라면 추가 불가
a.extend(100)    # 이러면 에러뜸

```python
a = [1, 2, 3]
a.extend([4, 5, 6])
print(a)
```

.pop() 또는 .pop(i)

* 리스트에서 지정한 인덱스의 항목을 제거하고 반환
* 작성하지 않을 경우 마지막 항목을 제거

```python
a = [1, 2, 3, 4, 5]

a_1 = a.pop()
a_2 = a.pop(0)

print(a_1)
print(a_2)
print(a)
```

** 리스트 정렬 및 순서 변경 뒤집기 메서드

.reverse()

* 리스트의 순서를 역순으로 변경(정렬X)

```python
a = [1, 2, 3, 4, 5]
a.reverse()
print(a.reverse)
print(a)
```

.sort()

* 원본 리스트를 오름차순으로 정렬

```python
a = [3, 4, 66, 22, 99]
a.sort()
print(a)
```

```python
a = [3, 4, 66, 22, 99]
a.sort(reverse = True)    # 내림차순 정렬
print(a)
```

가변/불변 객체의 개념

* 객체 복사의 핵심을 파악하려면, 파이썬 자료구조의 가변과 불변을 살펴봐야 함

Mutable(가변) 객체

* 생성 후 내용을 변경할 수 있는 객체
* list, dict, set
* 객체의 내용이 수정되어도 같은 메모리 주소 유지

Immutable(불변) 객체

* 생성 후 내용을 변경할 수 없는 객체
* int, float, str, tuple
* 새로운 값을 할당하면 새로운 객체 생성됨, 변수는 새 객체를 참조하게 됨
